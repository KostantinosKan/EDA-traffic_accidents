<h3><font color = "green">Data Preprocessing</font></h3>

The data will pass through the following processes and transformations:
<br></br>
* Lists of numerical boolean, and categorical (>2 categories) features.
* Convert boolean objects to numerical values 0 and 1.
* Convert Start_Time and End_Time to timestamps using pd.to_datetime.
* Add additional columns for starting hour, day_of_week, month and year, date.
* Add an additional column named "Count" equal to 1, for analysis reasons.

```python
# Numerical, boolean and categorical lists of features
bol_cols = [col for col in accidents_df.columns if accidents_df[col].dtype == object and accidents_df[col].nunique() == 2]
cat_cols = [col for col in accidents_df.columns if accidents_df[col].dtype == object and accidents_df[col].nunique() > 2]
num_cols = [col for col in accidents_df.columns if accidents_df[col].dtype != object]
```

```python
### Convert boolean objects to numerical values.

## We will change the 4 features with same categories ("Day", "Night") together and the feature "Side" seperately.

# Convert "Sunrise_Sunset", "Civil_Twilight", "Nautical_Twilight", and "Astronomical_Twilight" features
accidents_df[bol_cols[1:]] = accidents_df[bol_cols[1:]].replace({"Day": 1, "Night": 0})

# Convert "Side" feature
accidents_df["Side"] = accidents_df["Side"].replace({"R": 1, "L": 0})

# Check the changes
accidents_df[bol_cols].info()
```

```python
# Convert Star_Time to timestamps
accidents_df[["Start_Time", "End_Time"]] = accidents_df[["Start_Time", "End_Time"]].apply(pd.to_datetime)

# Add columns for starting hour, day of week, month and year
accidents_df["Start_Day_Week"] = accidents_df["Start_Time"].dt.dayofweek
accidents_df["Start_Month"] = accidents_df["Start_Time"].dt.month
accidents_df["Start_Date"] = accidents_df["Start_Time"].dt.date
accidents_df["Start_Hour"] = accidents_df["Start_Time"].dt.hour
accidents_df["Start_Year"] = accidents_df["Start_Time"].dt.year

# Add the column "Count"
accidents_df["Count"] = 1
```

```python
# Taking a sample
accidents_df = accidents_df.sample(750000)
accidents_df
```
